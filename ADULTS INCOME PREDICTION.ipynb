```python
countries = []
def get_country():
    for country in data2['native.country']:
        if country == "United-State": 
            country_name = "United States"
        else:
            country_name = country.split("-")[1] if "-" in country else country
        countries.append(country_name)
    return countries
countries = get_country()
```

```python
useful_columns=data2[['age','education','workclass','native.country','occupation','hours.per.week','income']]
```

```python
#  PERFORM TRANSFORMATION 
transformed_columns=useful_columns[['education','workclass','native.country','occupation']]
```

```python
transformed=pd.get_dummies(transformed_columns).astype(int)
transformed.head(3)
```

```python
data3=pd.concat([useful_columns,transformed],axis="columns")
```

```python
droped_columns=data3[['education_10th','education','workclass','native.country','occupation']]
data4=data3.drop(columns=droped_columns)
```

```python
data4.columns
```

```python
data4.head(2)
```

```python
# spliting the data and buid the model 
```

```python
x=data4.drop(columns="income")
y=data4.income
```

```python
# SPLITING THE DATA TO TEST 20% AND TRAIN 80%
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB,GaussianNB
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# passing the models and parameters
```

```python
models_parameters={
    "rain_forest":{
        'model':RandomForestClassifier(),
        'params':{}
    },
    "logistic":{
        'model':LogisticRegression(),
        'params':{
             'C': [10]
        }
    },
    "descion":{
        'model':DecisionTreeClassifier(),
        'params':{}
    }

    # "svm":{
    #     'model':SVC(),
    #     'params':{}
    # },

    # "mb":{
    #     'model':MultinomialNB(),
    #     'params':{}
    # },
    # "gb":{
    #     'model':GaussianNB(),
    #     'params':{}
    # }

}
```

```python
results=[]
for model,params in models_parameters.items():
    gs=GridSearchCV(params["model"],params["params"],return_train_score=False)
    gs.fit(X_train,y_train)
    results.append({
        'model':model,
        'best_score':gs.best_score_,
        'best_params':gs.best_params_
    })
df=pd.DataFrame(results,columns=['model','best_score','best_params'])   
```
