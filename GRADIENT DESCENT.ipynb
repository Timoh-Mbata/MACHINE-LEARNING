```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('insurance_data.csv')
data.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data.drop(columns="bought_insurance"),data.bought_insurance,test_size=0.25,random_state=25)
```

```python
X_train.shape
```

```python
X_train.columns
```

```python
X_train.age = X_train.age/100
X_test.age = X_test.age/100
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import warnings 
warnings.filterwarnings(action="ignore")
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    #keras.layers.Dense(20,input_shape = (2,), activation="relu"),
    #keras.layers.Dense(10,activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros'),
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5000)
```

```python
model.evaluate(X_test,y_test)
```
