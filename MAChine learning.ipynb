```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("(ML) LINEAR REGRESSION1.csv")
print(df.head())
```

```python
%matplotlib inline 
plt.scatter(df.AREA,df.PRICES)
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["AREA"]],df.PRICES)
```

```python

# Create a DataFrame for prediction
area_to_predict = pd.DataFrame({'AREA': [300]})

# Make a prediction
predicted_price = reg.predict(area_to_predict)

print(predicted_price)
```
