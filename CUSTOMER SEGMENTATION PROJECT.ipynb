```python
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1=pd.read_csv('Online_Retail_Data_Set[1].csv',encoding='ISO-8859-1', encoding_errors='replace')
data1.shape
```

```python
data2=pd.read_csv("Mall_Customers[1].csv")
data2.shape
```

```python
data2.columns
```

```python
data_2_1=data1[['Description', 'Quantity','UnitPrice','Country']]
data_2_1.shape
```

```python
data_2_3=data_2_1[:200]
data_2_3.shape
```

```python
data_2_2=data2[['Genre', 'Age', 'Annual Income (k$)','Spending Score (1-100)']]
```

```python
data=pd.concat([data_2_3,data_2_2],axis='columns')
data.shape
```
