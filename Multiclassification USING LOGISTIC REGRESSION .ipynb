```python
import pandas as pd
%matplotlib 
import matplotlib.pyplot as plt
import numpy as np
```

```python
'''
Use sklearn.datasets iris flower dataset to train your model using logistic regression.
You need to figure out accuracy of your model and use that to predict different samples in your test dataset.
In iris dataset there are 150 samples containing following features,
Sepal Length
Sepal Width
Petal Length
Petal Width

Using above 4 features you will clasify a flower in one of the three categories,
1.Setosa
2.Versicolour
3.Virginica
'''
```

```python
from sklearn.datasets import load_iris
```

```python
flower=load_iris()
# flower.data
```

```python
flower.target[]
```

```python
flower.target_names
```

```python
#
# import matplotlib.pyplot as plt
_, ax = plt.subplots()
scatter = ax.scatter(flower.data[:, 0], flower.data[:, 1], c=flower.target)
ax.set(xlabel=flower.feature_names[0], ylabel=flower.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], flower.target_names, loc="lower right", title="Classes"
)
```

```python
#  creating a model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# train my model 
x=flower.data
y=flower.target
X_train, X_test, y_train, y_test =train_test_split(x,y,train_size=0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
y_test
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict([[6.4, 3.2, 4.5, 1.5]])
```

```python
model.score(X_train,y_train)
```

```python
# practice on the confusion matrix
ypredicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,ypredicted)
confusion
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(confusion, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
