```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income[1].csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python

%matplotlib inline
x=df["year"]                       
y=df["per capita income (US$)"]
plt.scatter(x ,y, marker="+", color="r")
plt.plot(x ,y)
plt.title("CANADA PER CAPITA INCOME")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.grid()
plt.show()
```

```python
# CREATING A LINEAR MODEL
#  creating a linear regression Object
reg=linear_model.LinearRegression()
#  fit the data
# Extracting features and target variable
x = df[['year']]
y = df['per capita income (US$)']
# Fitting the model
reg.fit(x, y)
```

```python
x_new = pd.DataFrame({'year': [2025]})
# Making predictions
predictions = reg.predict(x_new)
# Output predictions
print(predictions)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
x=df["year"]                       
y=df["per capita income (US$)"]
plt.scatter(x ,y, marker="+", color="r")
plt.title("CANADA PER CAPITA INCOME")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.plot(df["year"] ,reg.predict(df[["year"]]),color="blue") 
plt.show()
```

```python
import joblib
joblib.dump(reg,"Canada per capita income module prediction")
```
