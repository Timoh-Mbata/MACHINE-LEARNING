```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
    # PERFORMING EDA 
```

```python
sepal_len=df['SepalLengthCm']
filter=df.groupby('Species')['SepalLengthCm'].sum()
x=filter.index
y=filter.values
color=['cyan','green','purple']
plt.figure(figsize=(8,6))
plt.bar(x,y, color=color)
plt.xlabel("TYPE OF SPECIES ")
plt.ylabel(" Sepal LENGTH IN cm ")
plt.title(" COMPARISON OF SPECIES AND THERE Sepal LENGTH in Cm ")
plt.show()
```

```python
sepal_len=df['SepalWidthCm']
filter=df.groupby('Species')['SepalWidthCm'].sum()
x=filter.index
y=filter.values
color=['cyan','green','purple']
plt.figure(figsize=(8,6))
plt.bar(x,y, color=color)
plt.xlabel("TYPE OF SPECIES ")
plt.ylabel("Sepal Width Cm ")
plt.title(" COMPARISON OF SPECIES AND THERE Sepal Width Cm ")
plt.show()
```
